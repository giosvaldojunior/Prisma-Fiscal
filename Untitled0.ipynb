```r
install.packages('rbcb')
install.packages('sidrar')
install.packages('tidyverse')
install.packages('lubridate')

# Carregue os pacotes
library(rbcb)
library(sidrar)
library(tidyverse)
library(lubridate)
```

```r
# Defina início
inicio <- "2015-12-01"

# ---- 1. IBC-BR (% mensal) ----
ibcbr <- get_series(code = 24364, start_date = "2015-12-01") %>%
  rename(data = date, ibcbr = '24364')

# ---- 2. Câmbio (média mensal R$/US$) ----
# 1. Baixar a série diária
cambio_diario <- get_series(code = 1, start_date = inicio) %>%
  rename(value = '1')

# 2. Obter último valor do mês (último dia útil com cotação)
cambio_mensal <- cambio_diario %>%
  mutate(mes = floor_date(date, "month")) %>%
  group_by(mes) %>%
  slice_max(order_by = date, n = 1) %>%
  ungroup() %>%
  rename(data = mes, cambio = value) %>%
  select(data, cambio)

# ---- 3. Expectativa IPCA (Focus - 12 meses à frente) ----
ipca_expec <- get_market_expectations("monthly", "IPCA", start_date = inicio)

# 1. Converter "MM/YYYY" para data real (primeiro dia do mês)
ipca_expec <- ipca_expec %>%
  mutate(
    DataReferencia = dmy(paste0("01/", DataReferencia)),
    Data = as.Date(Data)
  )

# 2. Calcular o horizonte em meses
ipca_expec <- ipca_expec %>%
  filter(baseCalculo == 0) %>%
  mutate(
    horizonte_meses = interval(Data, DataReferencia) %/% months(1),
    mes_coleta = floor_date(Data, "month")
  )

# 3. Filtrar para os horizontes desejados
ipca_expec_filtradas <- ipca_expec %>%
  filter(horizonte_meses %in% c(1, 3, 6, 12))

# 4. Para cada mês de coleta e horizonte, pegar a última expectativa disponível
ipca_expec_mensais <- ipca_expec_filtradas %>%
  group_by(mes_coleta, horizonte_meses) %>%
  filter(Data == max(Data)) %>%
  ungroup()

# 5. Reorganizar para formato wide
ipca_expec_wide <- ipca_expec_mensais %>%
  select(mes_coleta, horizonte_meses, Mediana) %>%
  pivot_wider(
    names_from = horizonte_meses,
    values_from = Mediana,
    names_prefix = "ipca_exp_"
  )

# 6. Filtrar a partir de dezembro de 2015
ipca_expec_wide <- ipca_expec_wide %>%
  filter(mes_coleta >= as.Date("2015-12-01")) %>%
  arrange(mes_coleta)

# Defina o vetor de datas
datas <- ipca_expec_wide$mes_coleta

# Criação das séries como objetos ts mensais
ipca_exp_1 <- ts(ipca_expec_wide$ipca_exp_1, start = c(2015, 12), frequency = 12)
ipca_exp_3 <- ts(ipca_expec_wide$ipca_exp_3, start = c(2015, 12), frequency = 12)
ipca_exp_6 <- ts(ipca_expec_wide$ipca_exp_6, start = c(2015, 12), frequency = 12)
ipca_exp_12 <- ts(ipca_expec_wide$ipca_exp_12, start = c(2015, 12), frequency = 12)

# ---- 4. IPCA mensal (%) - Tabela 1737 ----
ipca <- get_sidra(api = "/t/1737/n1/all/v/63/p/all/d/v63%202") %>%
  mutate(data = ymd(paste0(`Mês (Código)`, "-01")),
         ipca = as.numeric(str_replace(`Valor`, ",", "."))) %>%
  select(data, ipca) %>%
  filter(data >= ymd("2015-12-01"))

# ---- 5. Taxa de desemprego (%) - Tabela 6381 ----
desemprego <- get_sidra(api = "/t/6381/n1/all/v/4099/p/all/d/v4099%201") %>%
  mutate(data = ymd(paste0(`Trimestre Móvel (Código)`, "-01")),
         desemprego = as.numeric(str_replace(`Valor`, ",", "."))) %>%
  select(data, desemprego) %>%
  filter(data >= ymd("2015-12-01"))
```
